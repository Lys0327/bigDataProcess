```python
print('=====')
for x in range(1,6):
  print('x= %d'%(x))
  for y in range(1,6):
    print('y = %d'%(y))
  print('======')
```

```python
for x in range(1,10):
  for y in range(1,10):
    print('%d*%d=%2d '%(x,y,x*y))
  print()
```

```python
for x in range(1,10):
  for y in range(1,10):
    print('%d*%d=%2d '%(x,y,x*y),end='')
  print()
```

```python
for x in range(1,10):
  for y in range(1,10):
    print('%d*%d=%2d '%(y,x,x*y),end='')
  print()
```

```python
for x in range(1,10):
  for y in range(1,10):
    print('{}*{}={:2d}'.format(y,x,x*y),end=' ')
  print()
```
